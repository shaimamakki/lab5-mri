```python
# Load Silver features (from Phase 2)
silver_path = "abfss://lakehouse@lab5mri.dfs.core.windows.net/silver/tumor_features/"

silver_df = spark.read.parquet(silver_path)

print("Silver feature columns:")
print(silver_df.columns)
print("Total rows:", silver_df.count())
```

```python
# Identify columns
all_columns = silver_df.columns
non_feature_cols = ["image_id", "label"]
feature_cols = [c for c in all_columns if c not in non_feature_cols]

print("Feature columns:", len(feature_cols))
```

```python
pdf = silver_df.toPandas()
print(pdf.shape)
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(
    pdf,
    test_size=0.2,
    stratify=pdf["label"],
    random_state=42
)

print("Train rows:", train_df.shape)
print("Test rows:", test_df.shape)
```

```python
train_spark = spark.createDataFrame(train_df)
test_spark = spark.createDataFrame(test_df)

train_path = "abfss://lakehouse@lab5mri.dfs.core.windows.net/gold/train.parquet"
test_path = "abfss://lakehouse@lab5mri.dfs.core.windows.net/gold/test.parquet"

train_spark.write.mode("overwrite").parquet(train_path)
test_spark.write.mode("overwrite").parquet(test_path)

print("Saved:")
print(" -", train_path)
print(" -", test_path)
```
